**Conclusões:**

- É importante analisar cuidadosamente qual o melhor gráfico a ser utilizado para transmitir a informação de forma clara, de acordo com a disponibilização das informações;
- Depois de avaliar o melhor gráfico, é preciso pensar na escala para que nenhum detalhe seja suprimido. No caso do histograma, por exemplo, é necessário ajustar a escala (bins), para que os intervalos, que influenciam no desenho do gráfico, possam trazer alguma informação relevante. Os intervalos devem fazer com que a descrição da amostragem seja mais detalhada, aproximando esse gráfico a uma curva do processo análogo de uma integral;
- A ordem de visualização e disponibilização dos elementos também é importante na hora de transmitir uma informação;
- Atentar-se a paleta de cores, pois elas devem estar de acordo com o contexto e não podem poluir o projeto.
