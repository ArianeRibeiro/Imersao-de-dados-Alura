```
#renomear coluna g-0
dados.rename(columns={"g-0": "g0"},inplace=True)
```

```
#renomear coluna drogas
dados.rename(columns={"droga": "composto"}, inplace=True)
```

```
dados.head()
```

```
#frequencia tratamento D1
tD1 = dados[(dados.dose=="D1")& (dados.tratamento)]
```

```
tD1
```

```
#proporção tratamento D1
tD1["tratamento"].value_counts(normalize=True)
```

```
tD2 = dados[(dados.dose=="D2") & (dados.tratamento)]
```

```
tD2
```

```
#proporção tratamento D2
tD2["tratamento"].value_counts(normalize=True)
```

**Conclusões iniciais:** 

- As informações sobre os tipos de drogas estão anonimizadas;
- Os valores de g0 estão normalizados;
- Há uma discrepância entre as proporções de tratamento com droga e com controle. Essa discrepância se mantem filtrando o tratamento por doses;
- A coluna drogas traz informações sobre compostos.

```
compostos= dados["composto"].value_counts().index[:5]
```

```
compostos
```

```
dados.query("composto in @compostos")
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8,6))
ax=sns.countplot(x="composto", data=dados.query("composto in @compostos"))
ax.set_title("Top 5 compostos")
plt.show()
```
