Quanto mais pr√≥ximo de 1 e -1, mais correlacionadas est√£o as vari√°veis, e quanto mais pr√≥ximas de 0 menor correlacionadas est√£o. Pode-se visualizar na matriz de correla√ß√£o:

```
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

As correla√ß√µes podem indicar um poss√≠vel caso que seja pass√≠vel de uma examina√ß√£o mais detalhada, mas n√£o implicam na causalidade.

Matriz de correla√ß√£o das C√©lulas c:

```
# Compute the correlation matrix
corr_cel = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cel, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cel, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Agora, para analisar a matriz entre as express√µes genicas e as c√©lulas, ser√£o pegas as colunas de 0 a 50, g0 √† g50 e c0 √† c50, fazendo um corte e depois 'colando' uma na outra:

```
pri = dados.loc[:,'g0':'g50']
sec = dados.loc[:,'c0':'c50']
osdois = pri.join(sec)
osdois
```

Correlacionando as duas vari√°veis:

```
var = osdois.corr()
var2 = var.loc[:'g50','c0':]
var2
```

```
# Compute the correlation matrix
corr1 = var2

# Generate a mask for the upper triangle
#mask = np.triu(np.ones_like(corr1, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr1, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.xlabel('Culturas de c√©lulas')
plt.ylabel('Genes')
plt.title('Correla√ß√£o entre c√©lulas e genes')
plt.show()
```
