Quanto mais próximo de 1 e -1, mais correlacionadas estão as variáveis, e quanto mais próximas de 0 menor correlacionadas estão. Pode-se visualizar na matriz de correlação:

```
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

As correlações podem indicar um possível caso que seja passível de uma examinação mais detalhada, mas não implicam na causalidade.

Matriz de correlação das Células c:

```
# Compute the correlation matrix
corr_cel = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cel, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cel, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Para saber se o modelo está 'bem treinado', é preciso pegar uma amostra dos dados para que o modelo seja capaz de receber um valor desconhecido e chegar em um valor satisfatório. 
Vamos separar o conjunto de dados em treino e teste para verificar a qualidade do modelo. Geralmente, 20% ou 25% dos dados são reservados para teste e 80% treino, mas depende do volume de dados.

Como os resultados da regressão logística não foram satisfatórios, vamos tentar um modelo de árvore de decisão.

O g-385 é o nó raiz, a partir dele a árvore é dividida em dois lado, esses dois lados devem ser o mais homogêneo possível.
Agora podemos tentar melhorar o índice se acerto do modelo de predição com o aumento do número de níveis (max_depht):

```
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

teste = []
treino = []

for i  in range(1, 15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```
A medida que os treinos melhoram, os testes decaem, isso acontece porque quanto mais profundidade colocamos na árvore mais características são capturadas dos dados de treino. No entanto, o modelo fica viciado nos dados de treino que ele acaba não conseguindo generalizar para um resultado que ele desconhece
```

A medida que os resultados dos treinos melhoram, os resultados dos testes decaem, isso acontece porque quanto mais profundidade colocamos na árvore, mais características são capturadas dos dados de treino, e, por isso, o modelo fica viciado nos dados de treino e acaba não conseguindo generalizar para um resultado que ele desconhece.
Esse processo é conhecido como 'Overfitting', e uma forma de corrigir é criando várias árvores de decisão para pegar amostras aleatórias do banco de dados, com Random Forest:

```
#Criar uma coluna para informar quando tem drogas ou não em uma linha específica
pd.get_dummies(dados_combinados['tratamento'])
```

```
from sklearn.ensemble import RandomForestClassifier

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns = ['tratamento', 'dose', 'tempo'])
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_random_forest = RandomForestClassifier()
modelo_random_forest.fit(x_treino, y_treino)
modelo_random_forest.score(x_teste, y_teste)
```

Resultado um pouco mais satisfatório, mas o ideal é que seja mais próximo de 90%.
