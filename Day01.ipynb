```
#renomear todas as colunas retirando o "-"
colunas = []
dados2 = dados 
for i in dados2.columns:
    i = i.replace ("-","")
    colunas.append(i)
dados2.columns = colunas
                 
```

**Conclusões iniciais:** 

- O gráfico de pizza não funciona bem para visualização de várias informações;
- As informações sobre os tipos de drogas estão anonimizadas, possivelmente por questões éticas, para não influenciar na análise;
- Os valores de g0 são genes com valores que provavelmente estão normalizados;
- Há uma discrepância entre as proporções de tratamento com droga e com controle. Esse desbalanceamento se mantem filtrando o tratamento por doses. Isso acontece porque quando um experimento é realizado é necessário que se analize o comportamento das amostras submetidas a alguma substância em relação a amostras que estão em condições normais (com_controle), por isso, não é necessário um grande número de amostras em condições normais.
